```python
# embeddings = HuggingFaceEmbeddings()

embeddings = HuggingFaceEmbeddings(model_name="intfloat/multilingual-e5-large")

# embeddings = GPT4AllEmbeddings()

# embeddings = TensorflowHubEmbeddings()

# embeddings = HuggingFaceBgeEmbeddings()
```

```python
# 업로드하고 실행해야함
fn_dir = "/content/drive/My Drive/Colab Notebooks/data/terms"
```

```python
# loader = PyPDFDirectoryLoader(fn_dir) # PDF문서
loader = DirectoryLoader(fn_dir)
documents = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50) # Chunk 1000 단위(토큰)로 텍스트 자름 ## 자를 때 overlap
docs = text_splitter.split_documents(documents)

len(documents), len(docs)
```

```python
db = Chroma.from_documents(docs, embedding=embeddings,
                                 persist_directory=".")
db.persist()
```

```python
query = "당뇨병에 걸리면 어떤 증상이 있어?"
doc = db.similarity_search(query)
```
